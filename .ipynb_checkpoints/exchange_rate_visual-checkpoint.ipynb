# The Euro and the Dollar During the Financial Crisis of 2007–2008
## Objectives
Through this analysis, we are hoping to identify long-term trends in the Euro-US dollar exchange rate (EUR/USD) with a particular focus on the the financial crisis of 2007–2008, also known as the global financial crisis (GFC). We will seek to identify periods of heightened volatility and identify specific catalysts for extreme EUR/USD movements.

## Introducing the Dataset
This dataset consists of daily exchange rates (vs. the euro) between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. The data source is the European Central Bank (ECB) and was uploaded to [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) by Daria Chemkaeva. While the data are updated regularly, the following analysis is based on the dataset as of January 2021. For the purposes of this analysis, we will focus exclusively on the relationship between the euro and the US dollar.

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

## Data Cleaning
Below we rename columns, isolate the rows of interest (date and USD), and eliminate rows without the required data.

```python
# Rename relevant columns to align with our naming convention 
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar', 
                               'Period\\Unit:': 'Time'}, 
                      inplace=True)

exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
# EUR/USD is only relationship of interest
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.reset_index(inplace=True, drop=True)
euro_to_dollar.info()
```

## EUR/USD Relationship (Quoted as USD per 1 EUR)
#### Exchange Rates: Daily Values
First, we look at the relationship based on daily exchange rates. While informative, this chart features a lot of noise due to the high volume of data points.

```python
import matplotlib.pyplot as plt
# Enables Jupyter to display graphs
%matplotlib inline 

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

#### Exchange Rates: Rolling Averages
Below we display subplots with rolling averages of different horizons. Taking the rolling average allows us to identify long-term trends wihout the distraction of unwanted noise.

Below we show rolling 30-day, 60-day, 90-day, 180-day, and 360-day averages. Note that the rolling horizons do not line up with calendar dates since the dataset covers only business days.

```python
plt.figure(figsize=(9, 6))
plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], 
         euro_to_dollar['US_dollar'], 
         linewidth=1.5)
plt.title('Daily Values', weight='bold')
plt.yticks([0.5, 1, 1.5])
plt.ylim(0.4,1.7)

for i, rolling in zip(range(2, 7), [30, 60, 90, 180, 360]):
    plt.subplot(3, 2, i)
    plt.plot(euro_to_dollar['Time'], 
             euro_to_dollar['US_dollar']
             .rolling(rolling).mean(),
             linewidth=1.5)
    plt.yticks([0.5, 1, 1.5])
    plt.ylim(0.4,1.7)
    plt.title('Rolling {}-Day Averages'.format(rolling), 
              weight='bold')

plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

Based on these subplots, the 30-day rolling averages seems to strike the right balance of removing excess noise while providing sufficiently granual data. Based on this assessment, we will add rolling 30-day averages to the `euro_to_dollar` DataFrame.

```python
euro_to_dollar['rolling_mean_30'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

## EUR/USD Movement During the Financial Crisis of 2007–2008
#### Spikes in Volatility Correspond to Market Turbulence
Unsurprisingly, the most pronounced spikes in volatility are in late 2008 and early 2009.

```python
plt.figure(figsize=(9, 6))
plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar']
         .rolling(5).std(), 
         linewidth=1.5)
plt.title('Rolling 5 Day Std', weight='bold')
plt.yticks([0.05, 0.1])
plt.ylim(0,0.13)

for i, rolling in zip(range(2, 7), [30, 60, 90, 180, 360]):
    plt.subplot(3, 2, i)
    plt.plot(euro_to_dollar['Time'], 
             euro_to_dollar['US_dollar']
             .rolling(rolling).std(),
             linewidth=1.5)
    plt.yticks([0.05, 0.1])
    plt.ylim(0,0.13)
    plt.title('Rolling {} Day Std'.format(rolling), weight='bold')

plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

The spikes in the charts above align with the GFC. According to the [U.S. National Bureau of Economic Research](https://www.bls.gov/spotlight/2012/recession/) (the official arbiter of U.S. recessions) the recession began in December 2007 and ended in June 2009, and thus extended over eighteen months.

Below we segment `euro_to_dollar` into three distinct horizons: 
1. `pre_gfc` (Jan 1999 - Nov 2007)
2. `gfc` (Dec 2007 - Jun 2009)
3. `post_gfc` (Jul 2009 - Jan 2021).

```python
pre_gfc = euro_to_dollar.copy()[euro_to_dollar['Time'] < pd.Timestamp(2007, 12, 1)]
gfc = euro_to_dollar.copy()[(euro_to_dollar['Time'] >= pd.Timestamp(2007, 12, 1)) & 
                            (euro_to_dollar['Time'] <= pd.Timestamp(2009, 6, 30))]
post_gfc = euro_to_dollar.copy()[euro_to_dollar['Time'] > pd.Timestamp(2009, 6, 30)]
```

```python
import matplotlib.style as style

style.use('fivethirtyeight')
plt.figure(figsize=(12, 6))

ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 2)
ax3 = plt.subplot(2, 3, 3)
ax4 = plt.subplot(2, 1, 2)
axes = [ax1, ax2, ax3]

period = [pre_gfc, gfc, post_gfc]
# Different color ascribed to pre-GFC, GFC, post-GFC
color = ['#e2d810', '#d9138a', '#12a4d9']

for ax, p, c in zip(axes, period, color):
    ax.set_yticks([1, 1.2, 1.4, 1.6])
    ax.set_yticklabels([1.0, 1.2, 1.4, 1.6])
    ax.set_ylim(0.8, 1.7)
    ax.plot(p['Time'], p['rolling_mean_30'], color=c)

for p, c in zip(period, color):
    ax4.plot(p['Time'], p['rolling_mean_30'], color=c)
    
ax4.set_yticks([1, 1.2, 1.4, 1.6])
ax4.set_yticklabels([1.0, 1.2, 1.4, 1.6])
ax4.set_ylim(0.8, 1.7)


ax1.set_xlim(pd.Timestamp(2000, 1, 1))
ax1.set_xticks([pd.Timestamp(2000,1,1), pd.Timestamp(2001,1,1), 
                pd.Timestamp(2002,1,1), pd.Timestamp(2003,1,1), 
                pd.Timestamp(2004,1,1), pd.Timestamp(2005,1,1), 
                pd.Timestamp(2006,1,1), pd.Timestamp(2007,1,1)])
ax1.set_xticklabels(['2000', '', '2002', '','2004', '','2006', ''], 
                    rotation=45)
ax1.text(11000, 1.78,'Pre-GFC', weight='bold', color='#e2d810', 
         size=18)
ax1.text(11000, 1.68,'(Jan 2000 - Nov 2007)', weight='bold', 
         color='dimgrey', size=12)

ax2.set_xticks([pd.Timestamp(2008,1,1), pd.Timestamp(2009,1,1)])
ax2.set_xticklabels(['2008', '2009'], rotation=45)
ax2.text(13830, 1.78,'GFC', weight='bold', color='#d9138a', 
         size=18)
ax2.text(13830, 1.68,'(Dec 2007 - Jun 2009)', weight='bold', 
         color='dimgrey', size=12)

ax3.set_xticks([pd.Timestamp(2010,1,1), pd.Timestamp(2011,1,1), 
                pd.Timestamp(2012,1,1), pd.Timestamp(2013,1,1), 
                pd.Timestamp(2014,1,1), pd.Timestamp(2015,1,1), 
                pd.Timestamp(2016,1,1), pd.Timestamp(2017,1,1),
                pd.Timestamp(2018,1,1), pd.Timestamp(2019,1,1), 
                pd.Timestamp(2020,1,1)])

ax3.set_xticklabels(['2010', '', '2012', '', '2014', '', '2016', 
                     '', '2018', '', '2020'], rotation=45)
ax3.text(14260, 1.78,'Post-GFC', weight='bold', color='#12a4d9', 
         size=18)
ax3.text(14260, 1.68,'(Jan 2008 - Dec 2020)', weight='bold', 
         color='dimgrey', size=12)

ax4.set_xlim(pd.Timestamp(2000, 1, 1))
ax4.grid(axis='both', b=None)

# Add title and sub-header
ax1.text(11000, 2.09,'GFC Impact on EUR/USD exchange rate', 
         weight='bold', size=20)
ax1.text(11000, 1.97,'EUR/USD averaged 1.4 between Dec 2007 and Jun 2009', size=16)

plt.subplots_adjust(hspace = 0.3) # Ensures xtick labels aren't cut off
plt.show()
```

Takeaways:
1. From 2001 through the middle of 2008, EUR steadily appreciated vs. USD.
2. The average exchange rate (based on rolling 30-day averages) was significantly higher during the GFC (1.43) vs. pre-GFC (1.11) and post-GFC (1.23).
3. The most drastic change in rates took place in the second half of 2008.

To get a feel for specific events that triggered this volatility, we insert a new column into `euro_to_dollar` that quantifies one-day changes in the exchange rate.

```python
# No value for first date since calculation is n - (n-1)
one_day = [float('NaN')]
for i in range(1, euro_to_dollar.shape[0]):
    one_day.append(euro_to_dollar.loc[i]['US_dollar'] - 
                   euro_to_dollar.loc[i-1]['US_dollar'])
one_day_series = pd.Series(index=euro_to_dollar.index, data=one_day)
euro_to_dollar['one_day_change'] = one_day_series

euro_to_dollar
```

```python
euro_to_dollar.sort_values('one_day_change', ascending=True).head()
```

```python
euro_to_dollar.sort_values('one_day_change', ascending=True, 
                           na_position='first').tail()
```

#### Exchange Rate Swings between August 8, 2008 and March 19, 2009
Many of the biggest one-day swings (in each direction) were between August 8, 2008 and March 19, 2009. We will now deep dive into this period and diagnose some of the catalysts driving these significant one-day movements. Note that we are displaying daily values (as opposed to rolling averages) in the analysis below. Since we are analyzing a shorter horizon and focusing on one-day movements, we are less concerned with the noise in the data.

```python
fig, ax = plt.subplots(figsize=(12,6))
fig.set_facecolor('gainsboro')
ax.set_facecolor('white')
ax.font='arial'
ax.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'],
        color='#e23d28',linewidth=2)

# Display only left and bottom borders
for i in ['left', 'bottom']:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(1)
for i in ['top', 'right']:
    ax.spines[i].set_visible(False)
ax.tick_params(left=False, bottom=False)

ax.set_xticks([pd.Timestamp(2008, 8, 8), pd.Timestamp(2008, 9, 22),
               pd.Timestamp(2008, 10, 22), pd.Timestamp(2008, 12, 18), 
               pd.Timestamp(2009, 3, 19)])
ax.set_xticklabels(['Aug 8, 2008', 'Sep 22, 2008', 'Oct 22, 2008', 
                    'Dec 17-19, 2008', 'Mar 19, 2009'], 
                    size=12, color='black', weight='bold', rotation=45)
ax.set_xlim(pd.Timestamp(2008, 8, 1), pd.Timestamp(2009, 3, 26))
ax.set_yticks([1.2, 1.4, 1.6])
ax.set_yticklabels([1.2, 1.4, 1.6], size=12, color='black', weight='bold')
ax.set_ylim(1.1, 1.7)

ax.text(14096, 1.74, 'EUR/USD Reached Peak Volatility in Late 2008', 
        size=16, weight='bold')
ax.text(14096, 1.71, 'This span of seven months produced the largest one-day swings in exchange rates', 
        size=14)
ax.text(14094, 1.62, 'ECB president hints at weak economic growth', 
        size=8, wrap=True, rotation=345, backgroundcolor='mistyrose')
ax.text(14130, 1.32, '$700B govt. bailout plan', 
        size=8, wrap=True, backgroundcolor='mistyrose')
ax.text(14155, 1.2, 'USD/EUR reaches 20-month lows', 
        size=8, wrap=True, backgroundcolor='mistyrose')
ax.text(14195, 1.5, 'US Fed cuts short-term rates to historic lows', 
        size=8, wrap=True, backgroundcolor='mistyrose')
ax.text(14260, 1.15, 'US Fed announces massive buyback of bonds and MBS', 
        size=8, wrap=True, backgroundcolor='mistyrose', rotation=10)
ax.axvline(14099, ymin=.71, ymax=.85, linewidth=1, linestyle='--')
ax.axvline(14144, ymin=.40, ymax=.58, linewidth=1, linestyle='--')
ax.axvline(14174, ymin=.20, ymax=.31, linewidth=1, linestyle='--')
ax.axvline(14231, ymin=.60, ymax=.64, linewidth=1, linestyle='--')
ax.axvline(14322, ymin=.20, ymax=.40, linewidth=1, linestyle='--')
ax.grid(linewidth=.5)

plt.show()
```
